```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('basededados/monitoramento_tempo.csv', sep=',')
```

```python
import datetime
df['data']=pd.to_datetime(df['data'])
```

```python
df.describe()
```

```python
# Exercício 1
fig = plt.figure(figsize=(10, 8)) # Altera o tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # Está definindo os eixos
eixo.plot(df['data'],df['humidade'], color='red', lw=0.2, linestyle='--') # Adiciona os dados, modifica a cor, tamanho e estilo das linhas
eixo.legend(['Umidade'], loc='lower right', fontsize=15) # Define o nome, a posição e o tamanho da legenda
eixo.set_xlabel('Ano', fontsize=20) # O título do eixo x e seu tamanho
eixo.set_ylabel('Umidade', fontsize=20) # O título do eixo y e seu tamanho
```
